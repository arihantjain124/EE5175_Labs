Here i am going to reuse the gaussian kernel function defined in Lab 4 previously all this does is when a sigma is passed to it it builds a gaussian kernel with that sigma.

```python
def gaussiankernel(sigma):
    size=math.ceil(6*sigma+1)
    #size of kernel depending upon the sigma
    if size%2==0:
        size+=1
    kernel=np.zeros((size,size))
    #intializing kernel array with zeros
    if size==1:
        #mathematical operation fail for -ve value for sigma so added this check to account for it
        return np.array(([0, 0, 0],[0,1,0],[0, 0, 0]), dtype="int")
    temp1=1/(2*np.pi*(sigma**2))
    temp2=2*(sigma**2)
    #the function to calculate gaussian value for each cell have certain constant accross them so computing them beforehand
    with np.nditer(kernel, flags=['multi_index'], op_flags=['writeonly']) as it:
        #iterating over numpy array on both x and y axis together to compute the array efficently
        for h in it:
            m=it.multi_index[1]
            n=it.multi_index[0]
            h[...] = temp1*np.exp(-1*(((m-size/2)**2)+((n-size/2)**2))/temp2)
    kernel=kernel/np.sum(kernel)
    return kernel,size
```

Next we are going to import convolution function from Lab 4 again this applies a single kernel and slides it along the image passed to generate the output image.

```python
def convolution(kernel,img,pad):
    size=kernel.shape[0]
    img=np.pad(img, (pad, pad), 'constant')
    #Zero padding the image to work around edge cases in the image
    size=math.floor(size/2)
    res=np.zeros((img.shape[0]+size,img.shape[1]+size))
    #np array to store output array in
    for i in range(size,img.shape[0]-size):
        for j in range(size,img.shape[1]-size):
            #applying kernel on np array using slicing this allows us to reduce the length of code 
            # as well as use optimization done by numpy
            res[j][i] = (img[j-size:j+size+1,i-size:i+size+1]*kernel).sum()
    return res[size:img.shape[1]-size,size:img.shape[0]-size]
```

# Space variant blur

Space variant blur means sigma varies from point to point that means every pixel in the image has a sigma associated with it so the next function defined as sigmamap creates a new numpy array same size as image to hold sigma value for each pixel in the image. Later we will pass these value to the gaussian kernel function to get the kernel for that specific cell.

Sigmamap takes 2 argument

1.img  : To pass the image we want to generate the sigma map for 

2.func : To pass the function which tells how sigma varies accross the image.

The function given for this assignment is:

$$ \sigma (m,n) = A \exp \frac { -((m- \frac{N}{2})+ (n- \frac{N}{2}))} {B} $$

and we know for

$$ \sigma (0,0) = 0.01  $$


$$ \sigma (\frac{N}{2},\frac{N}{2}) = 2.00  $$

Using the equation we can find the value for A and B which comes out to be 

A=2 and B=8262.711 for N=195 which is the size of the input image we have

```python
def sigmamap(img,func):
    N=img.shape[0]
    win_size=[]
    a = np.arange(N)
    [x,y] = np.meshgrid(a,a)
    res = np.zeros((N,N))
    for i in range(N):
        temp=[]
        for j in range(N):
            res[i,j] = func(i,j,N)
            size=math.ceil(6*res[i,j]+1)
            #size of kernel depending upon the sigma
            if size%2==0:
                size+=1
            temp.append(size)
        win_size.append(temp)
    return res,win_size
```

The above function outputs the sigmamap which has a value for each pixel we are going to iterate over this in space-variant-blue function to process the image.

Below i have defined the function for space variant blur to be refered by sigmamap or any function.

```python
def func_1(m,n,N):
    a = 2
    b = 8262.711
    return (a*np.exp(-1*(((m-N/2)**2)+((n-N/2)**2))/b))
```

```python
def convolution_space_variant(img,func):
    sigmas,win_size=sigmamap(img,func)
    pad=np.max(win_size)//2
    padded_image = np.pad(img,(pad,pad),'constant')
    res=np.zeros_like(padded_image)
    for i in range(pad,img.shape[0]+pad):
        for j in range(pad,img.shape[1]+pad):
            x=i-pad
            y=j-pad
            kernel=gaussiankernel(sigmas[x][y])[0]
            size=win_size[x][y]//2
            res[i][j]=((padded_image[i-size:i+size+1,j-size:j+size+1])*kernel).sum()
    return res[pad:img.shape[0]+pad,pad:img.shape[1]+pad]
    
```

Above function takes in 2 parameter:
1.img : the image we want to apply space variant blur on
2.func: the function which describe variation of sigma throught the image

The function first calles the sigmamap function to generate the sigma map which is then iterated over in the for loop and a gaussian kernel is generated which is then used to find value for that pixel and this goes on for every pixel

```python
output=convolution_space_variant(globe,func_1)
```
