* Since the Image has 3 channel we need to create a function to add padding to all 3 channels of the image.

```python
def pad_RGB(image,pad):
    res_image=np.zeros((image.shape[0]+2*pad, image.shape[1]+2*pad, 3))
    for i in range(3):
        res_image[pad:-pad,pad:-pad,i]=image[:,:,i]
    return res_image
```

# PSNR calculation 

* This is the metric we will be using to see how significant have the difference is there between a good image and a noisy image.

* we compute PSNR in decibles and is given by 

$$ PSNR = 20 \log_{10} \frac {MAX_f} {\sqrt{MSE}} $$

Where $ MSR_f $ is the maximum intensity in the image and MSE is the mean-squared error between the good image and the noisy image.

```python
def PSNR(original_image, distorted_image):
    temp = np.mean((distorted_image-original_image)**2)
    return 20*np.log10(np.max(original_image)/np.sqrt(temp))
```

# Gaussian Weight for given 2 patches

* Given 2 Patches
* 1 Being the patch_i which we are checking for similarity in a given neighbourhood of cells.
* 2 patch_q is the window which is traversing over the neighbourhood.
* Each pixel for patch_q gets a weight based on how well it matches with the patch_i. This is computed using the gaussian approach which is defined as:

$$ G(p) = \frac {1}{Z} \Sigma_{q} G_\sigma ( p-q ) f(q) $$

```python
def gaussianweight(n_p,n_q,sigma):
    return  np.exp(np.sum((-1*(n_p-n_q)**2)/(2*sigma**2)))
```

```python
def NLM(patch_w,neighbourhood_w,image,h):
    d=patch_w+neighbourhood_w
    pad_image=pad_RGB(image,d)
    #padding the image to account for first few rows and columns
    sim_size=((2*patch_w)+1)//2
    w_size=((2*d)+1)//2
    res_image=np.zeros_like(image)
    #Creating a empty numpy array to store the results
    for i in range(res_image.shape[0]):
        for j in range(res_image.shape[1]):
            n_p = pad_image [ i-sim_size+d : i+sim_size+1+d , j-sim_size+d:j+sim_size+1+d,]
            # patch_i we are going to look into this patch neighbourhood assignment weight and then use
            # those weights as pixel weight for q
            Z=[0,0,0]
            
            for k in range(d):
                for l in range(d):
                    # iterating over the neighbourhood
                    x=i+neighbourhood_w+k
                    y=j+neighbourhood_w+l
                    n_q = pad_image[ x-sim_size:x+sim_size+1,y-sim_size:y+sim_size+1, ]    
                    # patch_q to hold the patch in iteration 
                    for rgb in range(3):
                        # iterating over each RGB channel and computing its weight acc. to gaussian function defined above
                        N_p=n_p[:,:,rgb].reshape(n_p[:,:,rgb].size)
                        N_q=n_q[:,:,rgb].reshape(n_q[:,:,rgb].size)
                        temp=gaussianweight(N_p,N_q,h)
                        Z[rgb]+=temp
                        res_image[i,j,rgb]+=temp*pad_image[x,y,rgb]
                        # assigning result pixel with gaussian weight
            res_image[i,j,:]=np.array((res_image[i,j,0]/Z[0],res_image[i,j,1]/Z[1],res_image[i,j,2]/Z[2]))
                        
    return res_image
                    

```

# Noise filtering using gaussian filters

* We can filter noise using gaussian filter we already have all the necessary function to implement this from previous assignments.

```python
def gaussiankernel(sigma,size):
    
    kernel=np.zeros((size,size))
    #intializing kernel array with zeros
    if size==1:
        #mathematical operation fail for -ve value for sigma so added this check to account for it
        return np.array(([0, 0, 0],[0,1,0],[0, 0, 0]), dtype="int")
    temp=2*(sigma**2)
    d=size//2
    #the function to calculate gaussian value for each cell have certain constant accross them so computing them beforehand
    for i in range(size):
        for j in range(size):
            kernel[i, j] = np.exp((-(i-d)**2-(j-d)**2)/(temp))
    kernel=kernel/np.sum(kernel)
    return kernel

def convolution_2d(kernel,img):
    size=kernel.shape[0]
    pad=size//2
    res=np.zeros((img.shape[0],img.shape[1]))
    img=np.pad(img, (pad, pad), 'constant')
    #Zero padding the image to work around edge cases in the image
    for i in range(res.shape[0]):
        for j in range(res.shape[1]):
            #applying kernel on np array using slicing this allows us to reduce the length of code 
            # as well as use optimization done by numpy
            res[i][j] = (img[i:i+(2*pad)+1,j:j+(2*pad)+1]*kernel).sum()
    return res


def gaussian_RGB_denoise(image,h):
    res_image=np.zeros_like(image)
    kernel=gaussiankernel(h,7)
    for i in range(3):
        res_image[:,:,i]=convolution_2d(kernel,image[:,:,i])
    return res_image
```

```python
h=0
psnr_gaussian=[]
res_gaussian=[]
x_axis=[]
for i in range(5):
    h+=0.1
    x_axis.append(h)
    res_gaussian.append(gaussian_RGB_denoise(krishna_0_001,h))
    temp=PSNR(krishna,res_gaussian[-1])
    psnr_gaussian.append(temp)
```

```python
plotter(res_gaussian)
```

* This is the output for Gaussian noise filtering.

* Here we can see clear difference between the 2 approach here the image still retains a rough texture which seems realistic. This is because of the fact that we are not concerating on the image itself.

* We are computing a gaussian kernel for a specific value of sigma and applying it this makes the filtering independent of the signal itself. This goes both ways.

* We can get better image representation where the noise is gaussian cause this would eliminate that noise more prominently than other kind of noise.

* This is a signal independent filter this means the image is not used as context which is what Non-Local Means is trying to do.

* The visual gives us a clear distinction between the 2 approaches and this is also obvious from the graph above. 
* Gaussian falls really slowly and its just trying to smooth the image unbiased on the underlying image. 
* This gives the gaussian a different curve then the other 2 approaches.

```python
def NLM_kernel(patch_w,neighbourhood_w,image,h,i,j):
    
    d=patch_w+neighbourhood_w
    pad_image=pad_RGB(image,d)
    sim_size=((2*patch_w)+1)//2
    w_size=((2*d)+1)//2
    kernel=np.zeros(([d,d]))
    
    n_p = pad_image [ i-sim_size+d : i+sim_size+1+d , j-sim_size+d:j+sim_size+1+d]
    
    Z=0
    count=0
    for k in range(d):
        for l in range(d):
            x=i+neighbourhood_w+k
            y=j+neighbourhood_w+l
            n_q = pad_image[ x-sim_size:x+sim_size+1,y-sim_size:y+sim_size+1]
            temp=gaussianweight(n_p,n_q,h)
            kernel[k,l]=temp
            Z+=temp
    return kernel
```
